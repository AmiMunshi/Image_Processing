```python
#import necessary packages
import numpy as np
```

```python
x= np.array([1,2,1,2]) # Enter the input sequence. 
```

```python
#Obtain forward transform
X= np.fft.fft(x) #using built in function from fft pack of numpy obtain the transformed sequence
print(X)
```

```python
# Obtain inverse transform
x_est= np.fft.ifft(X) #using built in function from fft pack for inverse trasnform
print(x_est)
```

```python
#Obtain the above twiddle factor matrix for forward transform
N=4
W= np.zeros((4,4), dtype= complex)
for n in range(0,N):
  for k in range(0,N):
    real= np.cos(2*np.pi*n*k/N)
    imaginary= -1*(np.sin(2*np.pi*n*k/N))
    W[n,k]= complex(real,imaginary)
print(W)

    
```

```python
X1= np.dot(W, np.transpose(x)) #Obtain transform without using the buil in function
print(X1)
```

```python
#Obtain the above twiddle factor matrix for inverse transform
N=4
W_inv= np.zeros((4,4), dtype= complex)
for n in range(0,N):
  for k in range(0,N):
    real= int(np.cos(2*np.pi*n*k/N))
    imaginary= int(np.sin(2*np.pi*n*k/N))
    W_inv[n,k]= complex(real,imaginary)
print(W_inv)
```

```python
#Apply inverse DFT on the transformed sequence X1
x1_est= np.dot(W_inv, X1)/N
print(x1_est)
```
