```python
#Power Law transformation
#s= c*r^g
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img= cv2.imread("/content/Fig0308(a)(fractured_spine).tif") #read the image
#plt.imshow(img,cmap="gray", vmin=0, vmax=255)
```

```python
plt.figure(figsize=(20,10)) #to set figure size
plt.subplot(1,4,1)
plt.imshow(img) #Plot the original image
#gamma g= 0.6
img_new1= np.array(255*(img / 255) ** 0.6, dtype = 'uint8') #power law formula
plt.subplot(1,4,2)
plt.imshow(img_new1, cmap="gray", vmin=0, vmax=255)
#gamma g= 0.4
img_new2= np.array(255*(img / 255) ** 0.4, dtype = 'uint8')
plt.subplot(1,4,3)
plt.imshow(img_new2, cmap="gray", vmin=0, vmax=255)
#gamma g= 0.3
img_new3= np.array(255*(img / 255) ** 0.3, dtype = 'uint8')
plt.subplot(1,4,4)
plt.imshow(img_new3, cmap="gray", vmin=0, vmax=255)
```
