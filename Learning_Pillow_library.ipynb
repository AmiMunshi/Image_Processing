```
# Image module
from PIL import Image
help(Image)
```

```
#Read and Write the image using PIL
file= "/content/lena_color_256.tif"
image= Image.open(file)
print(image) #Details of the image
```

```
#Show the image
image.show()
#Does not always works
```

```
#To dispaly the image
from IPython.display import display
display(image)
```

```
image1= image.copy() # Copy image using copy() function
image1.save("new_lena.tif") #save/write image using save() function
```

```
image1.convert("LA")
#https://pillow.readthedocs.io/en/stable/reference/Image.html
```

```
# Applying image filters
from PIL import ImageFilter
#help(ImageFilter)
```

```
image=image.convert('RGB') # this stands for red, green blue mode
blurred_image=image.filter(PIL.ImageFilter.BLUR)
display(blurred_image)
```

```
image_edge= image.filter(PIL.ImageFilter.FIND_EDGES)
display(image_edge)
```

```
image_contour= image.filter(PIL.ImageFilter.CONTOUR)
display(image_contour)
```

```
#Knowing height and widthe of the image
print(image.height, image.width)
#print("{}x{}".format(image.width, image.height))
```

```
#crop the image
help(image.crop)
display(image.crop((50,20,200,190)))
```

```
#Draw rectangle bounding box in the image
from PIL import ImageDraw
drawing_object=ImageDraw.Draw(image)
drawing_object.rectangle((50,20,200,190), fill = None, outline ='blue')
display(image)
```

```
# Image Enhancement using pillow
#Import image enhancement module
from PIL import ImageEnhance
#help(ImageEnhance)
```

```
#create a brightness object
enhancer=ImageEnhance.Brightness(image1)
#change brightness using enhance function
enhancer.enhance(0.5)
```

```
# generate images for different brightness level
images_bright=[]
for i in range(10):
  images_bright.append(enhancer.enhance(i/10))
#print(images_bright)
for i in range(len(images_bright)):
  display(images_bright[i])
```
