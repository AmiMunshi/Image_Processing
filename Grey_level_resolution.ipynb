Grey level resolution

```python
# Import cv2, matplotlib, numpy
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Read the original image and know its type
img1= cv2.imread('cameraman.png',0)
#type(img1)
#np.amax(img1)
```

```python
[m,n]=img1.shape
print(m,n)
```

```python
#Ask for the number of bit in which a pixel has to be represented in the new image
b= int(input("Enter the number of bits for the new image"))
```

```python
maxold= 256
maxnew= 2**b
constant = maxnew/maxold
```

```python
img2= np.zeros((m,n), dtype= np.int) #creating new image matrix

for i in range(m):
    for j in range(n):
        img2[i,j] = constant*img1[i,j]
```

```python
plt.imshow(img1,cmap="gray")
```

```python
plt.imshow(img2, cmap='gray')
cv2.imwrite("binary.png",img2)
```
